```python
import pandas as pd
import requests
import ast
```

```python
url = 'https://www.tgju.org/profile/geram18/performance'
```

```python
res = requests.get(url)
```

```python
begin_idx = res.text.find('$("#chart-6")')+len('$("#chart-6").msHighcharts({chartData:')+6
```

```python
# ast.literal_eval(res.text[begin_idx:])
len_idx = res.text[begin_idx:].find(']]')
list_str = res.text[begin_idx:begin_idx+len_idx+2]
list_of_prices = ast.literal_eval(list_str)
list_of_prices
```

```python
prices_list = list(map(lambda x: x[1], list_of_prices))
```

```python
from khayyam import JalaliDate
from datetime import timedelta
```

```python

start = JalaliDate(1403, 10, 19)
end = JalaliDate(1404, 4, 13)

dates = []
current = start
while current <= end:
    dates.append(current)
    current += timedelta(days=1)  # Add one day properly

for d in dates:
    print(d)
```
